```python
# Filling null values in feature with median of that feature
for feature in [feature for feature in df.columns if feature not in ['class']]:
  df[feature]=df[feature].fillna(df[feature].median())
```

```python
x=df.drop("class",axis = 1)
```

```python
y=df['class']
```

```python
y
```

```python
y = pd.get_dummies(y,drop_first=True)
y
```

```python
, train_test_split
#splitting the dataset into the Training set and testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size = 0.2,random_state = 0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense # dense layer is hidden layer.
from tensorflow.keras.layers import Dropout # dropout is is probability of select the neurons   kernel_initializer is weight instializer.
```

```python
classifier = Sequential()
classifier.add(Dense(units = 10, kernel_initializer = 'he_uniform',activation = 'relu',input_dim = 170))

classifier.add(Dense(units = 10, kernel_initializer = 'he_uniform',activation = 'relu'))

classifier.add(Dense(units = 1, kernel_initializer = 'glorot_uniform',activation = 'sigmoid'))
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy',metrics=['accuracy'])
```

```python
classifier.summary()
```

```python
model_history = classifier.fit(x_train,y_train,batch_size = 10, epochs =30,validation_split = 0.25)
```

```python
model_history.history.items()
```

```python
model_history.history['val_accuracy']
```

```python
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
```

```python
# list all data in history

print(model_history.history.keys())

# summarize history for accuracy
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc ='upper left')
plt.show()
```

```python
# summarize history for loss

plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'],loc ='upper left')
plt.show()
```

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred = y_pred>0.05
y_pred.shape
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from kerastuner.tuners import RandomSearch
from kerastuner.engine.hyperparameters import HyperParameters
```

```python
def build_model(hp):
  model=Sequential()
  model.add(Dense(units = 32,activation='relu',input_dim=170))
  model.add(Dense(units = 32,kernel_initializer = 'he_uniform',activation = 'relu'))
  model.add(Dense(1,activation='sigmoid'))
  optimizer = hp.Choice("optimizer",values=['adam','sgd','rmsprop','adadelta'])
  model.compile(optimizer= optimizer,loss='binary_crossentropy',metrics=['accuracy'])
  return model
```

```python
tuner = RandomSearch(build_model,objective='val_accuracy',max_trials=5,directory = 'project1',project_name = 'aps-fault')
```

```python
tuner.search(x_train,y_train,validation_data=(x_test,y_test),epochs=5)
```

```python
#hyperparmeter
from kerastuner.engine.hyperparameters import HyperParameters
from tensorflow.keras import optimizers
def build_model(hp):
  model = Sequential()
  for i in range(hp.Int('num_layers', 2, 20)): #number of layer 2 to 20.
      model.add(Dense(units = hp.Int('units_'+ str(i),
                                     min_value=32,
                                     max_value=512,
                                     step=32),
                      activation = 'relu'))
  model.add(Dense(1, activation ='sigmoid'))
  model.compile(optimizer = optimizers.Adam(hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])),loss='binary_crossentropy',metrics=['accuracy'])
  return model
 
```

```python
hp = HyperParameters()
```

```python
tuner_second = RandomSearch(build_model,objective='val_accuracy',max_trials=5,directory='project2',project_name='aps_faults')
```
